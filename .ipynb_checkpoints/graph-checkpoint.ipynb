```python
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib

## 基本

```python
x = np.linspace(-3, 3, 20)
y = x ** 2

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.show()
```

```python
# DataFrameの場合
import pandas as pd
df = pd.DataFrame(columns=['x', 'y'])
df.x = x
df.y = y
df
```

```python
x = np.linspace(-3, 3, 20)
y = x ** 2

plt.figure(figsize=(12, 8))
plt.plot(df['x'], df['y'])
plt.show()
```

```python
# indexを横軸にした描画
x = np.linspace(-3, 3, 20)
y = x ** 2

plt.figure(figsize=(12, 8))
plt.plot(df['x'])
plt.plot(df['y'])
plt.show()
```

```python
# 横軸を時系列にする
df.index = pd.date_range(start='2018-2-1', periods=20, freq='D')
```

```python
# indexを横軸にした描画(時系列)
x = np.linspace(-3, 3, 20)
y = x ** 2

plt.figure(figsize=(12, 8))
plt.plot(df['x'])
plt.plot(df['y'])
plt.show()
```

## subplots

<img src='http://ailaby.com/wp-content/uploads/2016/12/fig4.png'>

```python
y1 = x
y2 = x ** 2
 
# figure は 1 つ
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
 
axs[0, 0].plot(x, y1)
 
axs[0, 1].plot(x, y2)

axs[1, 0].plot(df['x'])
axs[1, 0].set_xticklabels(df.index, rotation=45)

axs[1, 1].plot(df['y'])
axs[1, 1].set_xticklabels(df.index, rotation=45)

plt.show()
```

## GridSpec

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
 
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
 
plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(2,3) #figureを2×3の領域に分割
 
plt.subplot(gs[0, :2]) #0行と0~1列を使う
plt.plot(x, y1)
plt.text(-2, 1.5, 'gs[0, :2]', size=15)
 
plt.subplot(gs[:, -1]) #全行と最後の列を使う
plt.plot(x, y2)
plt.text(-2, 8, 'gs[:, -1]', size=15)
 
plt.subplot(gs[1, 0]) #1行と0列を使う
plt.plot(x, y3)
plt.text(-2, 17, 'gs[1, 0]', size=15)
 
plt.subplot(gs[1, 1]) #1行と1列を使う
plt.plot(x, y4)
plt.text(-2, 70, 'gs[1, 1]', size=15)
 
plt.show()
```
